```python
# QUESTION 9. Write a function to calculate the mean of a list of numbers.

# So here if my list is not numbers then it will only check if the list is empty. If it it is it will return None. If I change my [1] with 'one' it will be false (not empty)
# then that means it will throw me an error.

def mean(numbers):
  return sum(numbers) / len(numbers) if numbers else None

numbers = [10, 20, 30, 68, 17, 1, 25, 60]
print(mean(numbers))
```

```python
#I can also use the inbuilt statistics function
import statistics

numbers = [10, 20, 30, 68, 17, 1, 25, 60]
numbers = statistics.mean(numbers)
print(numbers)
```
